# Load the dataset by importing the fashion_mnist using Keras

```python
import numpy as np
from keras.datasets import fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
# reshape dataset to have a single channel
x_train = x_train.reshape((x_train.shape[0], 28, 28, 1))
x_test = x_test.reshape((x_test.shape[0], 28, 28, 1)) 
```

```python
x_train, x_test
```

```python
import tensorflow as tf
tf.__version__
```

```python
len_train_images = len(x_train)
len_label_images=len(y_train)
len_test_images = len(x_test)
len_label_test_images=len(y_test)
print('Number of training Images in the Fashion MNIST Dataset',len_train_images)
print('Number of training Label Images in the Fashion MNIST Dataset',len_label_images)
print('Number of testing Images in the Fashion MNIST Dataset',len_test_images)
print('Number of testing label Images in the Fashion MNIST Dataset',len_label_test_images)
```

```python
x_train
```

```python
# Preprocessing the dataset before building a Neural Network

# normalize inputs (so all pixel values are transformed from [0,255] to [0,0-1.0]
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
x_train,x_test
```

```python
#num = len(np.unique(y_train))
num = len(y_train)
num
```

```python
y_train = np.array(y_train,dtype='int32')
y_test = np.array(y_test,dtype='int32')
```

```python
y_train,y_test
```

```python
# Encode the outputs with one hot coding
from keras.utils import np_utils
y_train= np_utils.to_categorical(y_train) #Converts a class vector (integers) to binary class matrix.
y_test = np_utils.to_categorical(y_test)
num_classes = y_test.shape[1]
```

# Build a neural network without convolutional layers to do the classification task 

```python
# Build Neural Network Layers
from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout
tf.set_random_seed(1)
np.random.seed(23)        
model = Sequential()
model.add(Flatten())
model.add(Dense(20, activation='relu')) 
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.add(Dense(num_classes, activation='softmax'))
```

```python
# Compile model
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
# fit the keras model on the dataset, default batch_size=32
epochs = 5
model.fit(x_train, y_train, epochs=epochs, batch_size=60)
```

```python
_, accuracy = model.evaluate(x_train, y_train)
print('Accuracy of the Model 1: %.2f' % (accuracy*100))
```

```python
_, accuracy = model.evaluate(x_test, y_test)
print('Testing Accuracy of the model 1: %.2f' % (accuracy*100))
```

```python
from keras.datasets import fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
# reshape dataset to have a single channel
x_train = x_train.reshape((x_train.shape[0], 28, 28, 1))
x_test = x_test.reshape((x_test.shape[0], 28, 28, 1))
```

```python
# Preprocessing the dataset before building a Neural Network

# normalize inputs (so all pixel values are transformed from [0,255] to [0,0-1.0]
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
# Encode the outputs with one hot coding
y_train= np_utils.to_categorical(y_train) #Converts a class vector (integers) to binary class matrix.
y_test = np_utils.to_categorical(y_test)
num_classes = y_test.shape[1]
```

# Build a neural network with the use of convolutional layers

```python
#TASK 2
from keras.constraints import maxnorm
from keras.optimizers import SGD
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
model_CNN = Sequential()
model_CNN.add(Conv2D(128, (3, 3), input_shape=(28, 28, 1), padding='same', activation='relu', kernel_constraint=maxnorm(3)))
model_CNN.add(Conv2D(128, (3, 3), activation='relu', padding='same', kernel_constraint=maxnorm(3)))
model_CNN.add(MaxPooling2D())
model_CNN.add(Flatten())
model_CNN.add(Dense(512, activation='relu', kernel_constraint=maxnorm(3)))
model_CNN.add(Dropout(0.2))
model_CNN.add(Dense(num_classes, activation='softmax'))
```

```python
# Define optimizer
lrate = 0.002
epochs = 5
decay = lrate/epochs
sgd = SGD(lr=lrate, momentum=0.7, decay=decay, nesterov=False) #Stochastic gradient descent optimizer

# Compile model
model_CNN.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
x_train.shape
```

```python
model_CNN.summary()
```

```python
tf.set_random_seed(1)
np.random.seed(23)
# fit the keras model on the dataset, default batch_size=32
epochs = 5
model_CNN.fit(x_train, y_train, epochs=epochs, batch_size=60)
```

```python
_, accuracy = model_CNN.evaluate(x_train, y_train)
print('Accuracy of the CNN Model 1: %.2f' % (accuracy*100))
```

```python
# Final evaluation of the model
scores = model_CNN.evaluate(x_test, y_test)
print("Testing Accuracy of the CNN Model 1: %.2f%%" % (scores[1]*100))
```

```python
# Task 3 
from keras.datasets import fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
# reshape dataset to have a single channel
x_train = x_train.reshape((x_train.shape[0], 28, 28, 1))
x_test = x_test.reshape((x_test.shape[0], 28, 28, 1)) 
```

```python
# Preprocessing the dataset before building a Neural Network

# normalize inputs (so all pixel values are transformed from [0,255] to [0,0-1.0]
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
y_train.shape
```

```python
# Encode outputs
y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)
num_classes = y_test.shape[1]
y_train.shape
```

# Change the type of optimizer or learning rate that you applied in the previous tasks

```python
from keras.optimizers import Adam

# Build a deeper CNN model
model_CNN_opt = Sequential()
model_CNN_opt.add(Conv2D(64, (3, 3), input_shape=(28, 28, 1), activation='relu', padding='same'))
model_CNN_opt.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model_CNN_opt.add(MaxPooling2D())
model_CNN_opt.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model_CNN_opt.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model_CNN_opt.add(MaxPooling2D())
model_CNN_opt.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
model_CNN_opt.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
model_CNN_opt.add(MaxPooling2D())
model_CNN_opt.add(Flatten())
model_CNN_opt.add(Dense(1024, activation='relu', kernel_constraint=maxnorm(3)))
model_CNN_opt.add(Dense(512, activation='relu', kernel_constraint=maxnorm(3)))
model_CNN_opt.add(Dropout(0.2))
model_CNN_opt.add(Dense(num_classes, activation='softmax'))
# Compile model
epochs = 20
lrate = 0.003
adam = Adam(lr=lrate)
model_CNN_opt.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
model_CNN_opt.summary()
# Fit the model
model_CNN_opt.fit(x_train, y_train, epochs=epochs, batch_size=64)
```

```python
# Final evaluation of the model
_, accuracy = model_CNN_opt.evaluate(x_train, y_train, verbose=0)
print('Accuracy of the CNN Model Optimizer : %.2f' % (accuracy*100))
scores = model_CNN_opt.evaluate(x_test, y_test, verbose=0)
print("Testing Accuracy of the CNN Model Optimizer : %.2f%%" % (scores[1]*100))
```
